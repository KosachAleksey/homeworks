Описание: В этой задаче необходимо создать граф, моделирующий систему
доставки товаров. Каждая вершина графа будет представлять склад или
магазин, а ребра — это дороги между ними. Веса ребер будут отражать
стоимость доставки между двумя точками (например, в деньгах или времени).
Задача состоит в том, чтобы с помощью библиотеки networkx найти самый
дешевый маршрут для доставки товара.
Требования
 Создание графа: 
 Создайте граф, где вершины будут представлять склады и магазины.
Добавьте не менее 5 вершин, а также 6-8 дорог между ними, каждая дорога
будет иметь свой вес, который может обозначать стоимость доставки
(например, в денежном эквиваленте)a
 Поиск минимальной стоимости маршрута: 
 Используя библиотеку networkx, найдите путь с наименьшей стоимостью
доставки от одного склада до магазина. Выведите сам путь и его общую
стоимостьa
 Визуализация графа: 
 Постройте граф, на котором будут отображаться склады и магазины, а
также дороги между ними. Подпишите на графе веса (стоимости) реберa
 Дополнительная задача (по желанию): 
 Добавьте новый склад или магазин в граф и пересчитайте минимальную
стоимость маршрута. Проанализируйте, как добавление нового узла
влияет на стоимость доставки.

```python
import networkx as nx
import matplotlib.pyplot as plt

# Создание графа
G = nx.Graph()

# Добавление вершин (складоов и магазинов)
nodes = ["Склад A", "Склад B", "Магазин 1", "Магазин 2", "Магазин 3"]
G.add_nodes_from(nodes)

# Добавление рёбер (дорог) с весами (стоимость доставки)
edges = [
    ("Склад A", "Магазин 1", 5),
    ("Склад A", "Магазин 2", 3),
    ("Склад A", "Склад B", 4),
    ("Склад B", "Магазин 2", 2),
    ("Склад B", "Магазин 3", 6),
    ("Магазин 1", "Магазин 2", 1),
    ("Магазин 2", "Магазин 3", 3),
]
G.add_weighted_edges_from(edges)

# Поиск кратчайшего пути от "Склад A" до "Магазин 3"
start = "Склад A"
end = "Магазин 3"
shortest_path = nx.shortest_path(G, source=start, target=end, weight="weight")
shortest_path_cost = nx.shortest_path_length(G, source=start, target=end, weight="weight")

print(f"Кратчайший путь от {start} до {end}: {shortest_path}")
print(f"Общая стоимость доставки: {shortest_path_cost}")

# Визуализация графа
pos = nx.spring_layout(G)  # Позиции для всех узлов

# Рисуем узлы
nx.draw_networkx_nodes(G, pos, node_size=700, node_color="lightblue")

# Рисуем рёбра
nx.draw_networkx_edges(G, pos, width=2, edge_color="gray")

# Подписываем узлы
nx.draw_networkx_labels(G, pos, font_size=12, font_family="sans-serif")

# Подписываем веса рёбер
edge_labels = nx.get_edge_attributes(G, "weight")
nx.draw_networkx_edge_labels(G, pos, edge_labels=edge_labels)

plt.title("Графф системы доставки товаров")
plt.axis("off")  # Отключаем оси
plt.show()

# Дополнительная задача: добавление нового склада и пересчёт
print("\nДополнительная задача: добавлениие нового слада")

# Добавляем новый склад и рёбра
G.add_node("Склад C")
new_edges = [
    ("Склад C", "Склад B", 3),
    ("Склад C", "Магазин 3", 2),
]
G.add_weighted_edges_from(new_edges)

# Ищем новый кратчайший путь
new_start = "Склад A"
new_end = "Магазин 3"
new_shortest_path = nx.shortest_path(G, source=new_start, target=new_end, weight="weight")
new_shortest_path_cost = nx.shortest_path_length(G, source=new_start, target=new_end, weight="weight")

print(f"Новый кратчайший путь от {new_start} до {new_end}: {new_shortest_path}")
print(f"Новая общая стоимость доставки: {new_shortest_path_cost}")

# Обновлённая визуализация
plt.figure()  # Создаём новое окно для графа
new_pos = nx.spring_layout(G)  # Обновляем позиции с новым узлом

nx.draw_networkx_nodes(G, new_pos, node_size=700, node_color="lightblue")
nx.draw_networkx_edges(G, new_pos, width=2, edge_color="gray")
nx.draw_networkx_labels(G, new_pos, font_size=12, font_family="sans-serif")
new_edge_labels = nx.get_edge_attributes(G, "weight")
nx.draw_networkx_edge_labels(G, new_pos, edge_labels=new_edge_labels)

plt.title("Обновлённый граф с новым складом")
plt.axis("off")
plt.show()
```
