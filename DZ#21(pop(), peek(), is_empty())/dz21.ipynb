Описание: В этом задании вам необходимо создать реализацию стека на
языке Python. Стек должен принимать только один тип данных, который
будет указан при его создании. Важно, чтобы операции с данным стеком
(добавление элемента, извлечение элемента и другие) проверяли, что
тип добавляемого или извлекаемого элемента соответствует
указанному типу при создании стека.

 Создайте класс Stack, который реализует стекg
 Конструктор класса должен принимать параметр data_type,
который будет определять тип данных, который стек может
содержать (например, int, str и т. д.)g
 Стек должен поддерживать следующие операцииv
 push(item): добавление элемента в стек. Если тип элемента не
соответствует типу, указанному при создании стека, нужно

 pop(): извлечение элемента из стека. Если стек пуст,

 peek(): просмотр верхнего элемента стека без его удаления.

Gg Проверка типа данныхv

типу, заданному при создании стекаg
 Если тип данных элемента не совпадает с типом, указанным при

```python
class Stack:
    def __init__(self, data_type):
       
        self.data_type = data_type
        self.stack = []
    
    def push(self, item):
       
        if not isinstance(item, self.data_type):
            raise TypeError("Неверный тип данных для стека")
        self.stack.append(item)
    
    def pop(self):
      
        if self.is_empty():
            raise IndexError("Стек пуст")
        return self.stack.pop()
    
    def peek(self):
       
        if self.is_empty():
            raise IndexError("Стек пуст")
        return self.stack[-1]
    
    def is_empty(self):
        
        return len(self.stack) == 0

# Тестирование
if __name__ == "__main__":
    # Создание стека только для int
    stack = Stack(int)
    
    # Добавление элементов
    stack.push(5)
    stack.push(10)
    print("Верхний элемент стека:", stack.peek())  # Должно вывести 10
    
    # Извлечение элементов
    print("Извлеченный элемент:", stack.pop())  # Должно вывести 10
    print("Стек пуст?", stack.is_empty())  # Должно вывести False
    print("Извлеченный элемент:", stack.pop())  # Должно вывести 5
    print("Стек пуст?", stack.is_empty())  # Должно вывести True
    
    # Проверка ошибок
    try:
        stack.pop()
    except IndexError as e:
        print("Ошибка:", e)  # Должно вывести "Стек пуст"
    
    try:
        stack.push("string")  # Попытка добавить строку в стек для int
    except TypeError as e:
        print("Ошибка:", e)  # Должно вывести "Неверный тип данных для стека"
```
