^ Напишите программу, которая выводит все чётные числа от 1 до 20
(включительно) с помощью цикла for
^ Затем напишите программу, которая будет запрашивать ввод чисел у
пользователя, пока пользователь не введёт 0. Все введённые числа

Ответ на 1 вопрос (Теория) Задания 3: В цикле for известно количество итераций, а значит определено, когда он закончится. В цикде while количество итераций не определено. Их количество зависит от того, какие были введены значения и от желания остановить цикл на любом этапе, вызовом команды break. В примере «Повторяю упражнение 10 раз испльзовать нужно цикл for, потомучто известно количество упражнений - 10 раз.  

```python
#ответ на 2 вопрос (Практическая часть) Задания 3
#программa, которая выводит все чётные числа от 1 до 20 (включительно) с помощью цикла for
def even():
    evens = []
    for num in range(1, 21):
        if num % 2 == 0:
            evens.append(num)
    return evens

even()

```

```python
#программа, которая запрашивает ввод чисел у пользователя, пока пользователь не введёт 0. Все введённые числа(кроме нуля) должны суммироваться, и в конце программа выводить сумму
sum_numbers = 0
while True:
    num = int(input("Введите число для суммирования: , если хотите прекратить нажимте 0"))
    if num == 0:
        break

    sum_numbers += num

print(f"Сумма всех веденных чисел равна: {sum_numbers}")
```

Лямбда-функции — это небольшие функции без имени, которые можно определить прямо в одном выражении. Они применяются, когда нужно быстро выполнить простую операцию, состоящую из одного выражения. Часто их используют для передачи функций в другие функции, такие как map, filter или reduce. Главная особенность лямбда-функций - они однострочны и ограничены одним выражением.

```python
#ответы на 2 вопрос (Практическая часть) Задания 8
result = lambda x: x ** 3
list_numbers = [1, 2, 3, 4, 5]
cubed_list = list(map(result, list_numbers))
print(cubed_list)
```

Задание 9. Callback-функциk
1. Теория
Объясните, что такое callback-функция (функция обратного вызова).
Приведите пример, где функция передаётся в другую функцию в
качестве аргумента, а затем вызывается внутри неё
2. Практическая часть
Напишите функцию processor(func, data), которая принимает функцию
func (это и есть callback) и список data.
Внутри processor вызовите переданную функцию func для каждого
элемента списка data и сохраните результат в новом спискеx.
Верните этот новый список из processor.
Создайте несколько простых функций (например, удваивание числа,
возведение в квадрат) и передайте их в processor.

Callback-функция  — это функция, которая передается в другую функцию в качестве аргумента и выполняется внутри этой внешней функции. Внешняя функция выполняет какие-то действия, а затем вызывает переданную ей функцию.

```python
#ответы на 2 вопрос (Практическая часть) Задания 9
def processor(func, data):
    new_list = []
    for item in data:
        new_list.append(func(item))
    return new_list
def double(x):
    return x * 2
def square(x):
    return x ** 2
data = [1, 2, 3, 4, 5]
doubled_data = processor(double, data)
squared_data = processor(square, data)
print("Удвоенные данные:", doubled_data)
print("Возведённые в квадрат данные:", squared_data)
```

```python
#ответы на 2 вопрос (Практическая часть) Задания 10
with open('data.txt', 'w', encoding='utf-8') as file:
    file.write('Файл открывается с помощью функции open(), которая принимает два основных параметра: путь к файлу и режим открытия\n')
    file.write('Режимы открытия файлов:\n')
    file.write('r – чтение, w – запись, a – добавление, а также rb, wb, ab и r+, w+, a+ \n')

# Чтение из файла
with open('data.txt', 'r', encoding='utf-8') as file:
    content = file.read()
    print(content)
```
