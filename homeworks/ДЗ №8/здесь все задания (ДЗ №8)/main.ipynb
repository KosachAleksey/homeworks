Задание 1
Пользователь вводит с клавиатуры строку. Проверьте 
является ли введенная строка палиндромом. Палинслева направо и справа налево. Например, кок; А роза упала на лапу Азора; доход; А буду я у дуба

```python
str =  input("Вводит с клавиатуры строку. Проверьте является ли введенная строка палиндромом. (все буквы должны вводиться в одном регистре )")
str = str.upper()
print("Слева направо: ", str[::1])
print("Справа налево: ",str[::-1])
if str[::1] == str [::-1]:
    print("Введенная строка является палиндромом")
elif str[::1] != str [::-1]:
    print("Введенная строка не является палиндромом")
```

Задание 2
Пользователь вводит с клавиатуры некоторый текст, 
после чего пользователь вводит список зарезервированных 
слов. Необходимо найти в тексте все зарезервированные 
слова и изменить их регистр на верхний. Вывести на 
экран измененный текст.

```python
text = input("Введите некоторый текст.")
reserv_words = input("Введите список зарезервированных слов через пробел: ").split()
words = text.split()
for i in range (len(words)): 
    if words[i] in reserv_words:
        words[i]=words[i].upper()
united_text = ' '.join(words)
print(f"""Измененный текст: "{united_text}" """)
```
